```python
pokedex=pd.read_csv('pokedex.csv')
```

```python
pokedex.head()
```

```python
bulbasaur=pokedex[pokedex.Name=='Bulbasaur'][['Name','Attack','Defense','SpecialAttack','SpecialDefense','Speed']].set_index('Name')
bulbasaur
```

```python
app = dash.Dash(__name__) 

app.layout=html.Div([
    html.H1(id='Header',children=['Pokemon Stat Checker'],style={'textAlign': 'center'}),
    dcc.Graph(id='bar',figure={
            'data': [{'x': [col for col in bulbasaur.columns], 'y': [stat for stat in bulbasaur.loc['Bulbasaur']], 'type': 'bar'}],
            'layout':{'title':'My Bar Chart'}
    })
])

application = app.server
if __name__ == '__main__':
    application.run(debug=False)
```

```python
app = dash.Dash(__name__) 

app.layout=html.Div([
    html.H1(id='Header',children=['Pokemon Stat Checker'],style={'textAlign': 'center'}),
    dcc.Graph(id='bar'),
    dcc.Dropdown(id='drop',
                    options=[{'label': i, 'value': i} for i in pokedex.Name],
                    #value='Bulbasaur',
                    #searchable=True,
                    style={'display': 'inline-block',
                          'width': '49%'}
                )
])

@app.callback(Output('bar','figure'),
            [Input('drop','value')])
def figure_update(value):
    pokemon=pokedex[pokedex.Name==value][['Name','Attack','Defense','SpecialAttack','SpecialDefense','Speed']].set_index('Name')
    figure={'data': [{'x': [col for col in pokemon.columns], 'y': [stat for stat in pokemon.loc[value]], 'type': 'bar'}],
            'layout':{'title':value}
    }
    return figure
        

application = app.server
if __name__ == '__main__':
    application.run(debug=False)
```

```python
app = dash.Dash(__name__) 

app.layout=html.Div([
    html.H1(id='Header',children=['Pokemon Stat Checker'],style={'textAlign': 'center'}),
    html.Div(children=[
    dcc.Graph(id='bar',style={'display':'inline-block'}
    ),
    dcc.Graph(id='pie',style={'display':'inline-block'})
    ]),
    dcc.Dropdown(id='drop',
                    options=[{'label': i, 'value': i} for i in pokedex.Name],
                    value='Bulbasaur',
                    #searchable=True,
                    style={'display': 'inline-block',
                          'width': '49%'}
                )
],
)

@app.callback([Output('bar','figure'),
               Output('pie','figure')],
            [Input('drop','value')])
def figure_update(value):
    pokemon=pokedex[pokedex.Name==value][['Name','Attack','Defense','SpecialAttack','SpecialDefense','Speed']].set_index('Name')
    total=pokedex[pokedex.Name==value].reset_index().loc[:,'Total'][0]
    bar={'data': [{'x': [col for col in pokemon.columns], 'y': [stat for stat in pokemon.loc[value]], 'type': 'bar'}],
            'layout':{'title':value}
    }
    pie=go.Figure(data= go.Pie(title='Stat Proportion',
                                      direction='clockwise',
                                      labels=[col for col in pokemon.columns],
                                      values=[pokemon[[col]].loc[value][0]/total for col in pokemon.columns],
                                      marker= {'colors': ['red','blue','yellow','green','purple']}
                                      
                                     ))
    return bar, pie
        

application = app.server
if __name__ == '__main__':
    application.run(debug=False)
```

```python
app = dash.Dash(__name__) 

app.layout=html.Div([
    html.H1(id='Header',children=['Pokemon Stat Checker'],style={'textAlign': 'center'}),
    html.Div(children=[
    dcc.Graph(id='bar'
    )
    ]),
    dcc.Dropdown(id='drop',
                    options=[{'label': i, 'value': i} for i in pokedex.Name],
                    value='Bulbasaur',
                    #searchable=True,
                    
                ),
    dcc.RadioItems(id='two',
                   options=[{'label':'Compare Off','value':1},{'label':'Compare On','value':2}],
                   value=1,
               ),
    dcc.Dropdown(id='drop2',style={'display':'none'},options=[{'label': i, 'value': i} for i in pokedex.Name],value='Squirtle')
],
)
@app.callback(Output('drop2','style'),
              [Input('two','value')])
def drop_two(value):
    if value==2:
        return {'display': True}
    else:
        return {'display':'none'}
@app.callback(Output('bar','figure'),
            [Input('drop','value'),
             Input('two','value'),
             Input('drop2','value')])
def figure_update(value,compare,two):
    pokemon=pokedex[pokedex.Name==value][['Name','Attack','Defense','SpecialAttack','SpecialDefense','Speed']].set_index('Name')
    if compare==1:
        bar={'data': [{'x': [col for col in pokemon.columns], 'y': [stat for stat in pokemon.loc[value]], 'type': 'bar'}],
                'layout':{'title':value}
        }
        return bar
    else:
        pokemon2=pokedex[pokedex.Name==two][['Name','Attack','Defense','SpecialAttack','SpecialDefense','Speed']].set_index('Name')
        fig = go.Figure(data=[
            go.Bar(name=value, x=[col for col in pokemon.columns], y=[stat for stat in pokemon.loc[value]]),
            go.Bar(name=two, x=[col for col in pokemon.columns], y=[stat for stat in pokemon2.loc[two]])
        ])
        return fig
        

application = app.server
if __name__ == '__main__':
    application.run(debug=False)
```
